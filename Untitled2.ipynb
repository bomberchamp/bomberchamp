```python
data = pickle.load( open( "train_beginning.p", "rb" ) )
X = data['X']
Y = data['Y']
```

```python

choices = ['RIGHT', 'LEFT', 'UP', 'DOWN', 'BOMB', 'WAIT']

K.clear_session()

D = len(choices)

#========================
#  Define Model
#========================
    
inputs = Input(shape=(s.cols, s.rows, 6))
x = Conv2D(8, 3, activation='relu')(inputs)
x = Conv2D(16, 3, activation='relu')(x)
x = Conv2D(32, 3, activation='relu')(x)
x = Dropout(0.3)(x)
x = Conv2D(64, 3, activation='relu')(x)
x = Dropout(0.3)(x)
x = Flatten()(x)
x = Dense(64, activation='relu')(x)
pred = Dense(D, activation='softmax')(x)

model = Model(inputs=inputs, outputs=pred)
model.compile(loss="categorical_crossentropy", optimizer="adam")
```

```python
perm = np.random.permutation(5000)
train = perm[:4000]
b = np.zeros((4000, D))
b[np.arange(4000), Y[train]] = 1

model.fit(X[train], b, epochs=100)
```

```python
test = perm[4000:]
b = np.zeros((1000, D))
b[np.arange(1000), Y[test]] = 1

model.evaluate(X[test], b)
```
